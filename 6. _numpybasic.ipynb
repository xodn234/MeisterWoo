### Numpy
- 데이터는 행열로 표현
- 행열 데이터 빠르게 계산을 해야 합니다.
- 행열 데이터 생성, 수정, 계산등의 빠르게 처리해주는 패키지
    - 특정
        - C, C++ 포트란로 작성
        - 선형대수학을 빠르게 연산
            - 스칼라, 벡터, 매트릭스

```python
import numpy as np
```

```python
# 행렬 데이터 생성 
# ndarray : 한가지 데이터 타입만 값으로 사용이 가능
```

```python
arrary = np.array([1, 2, 3])
type(arrary), arrary
```

```python
arrary2 = np.array(
    [[1, 2, 3],
    [4, 5, 6]]
)
arrary2, arrary2.ndim, arrary2.shape
```

```python
# 행렬의 모양(shape) 변경3, 2, 1기
arrary2.reshape(3, 2), arrary2.reshape(6), arrary2.reshape(3, 2, 1)
```

```python
# 행렬 데이터의 선택 : offset index : masking
```

```python
arrary2[1], arrary2[1][2], arrary2[1, 2], arrary2[1][::-1]
```

```python
# 데이터 수정
```

```python
ls = [1, 2, 3]
ls[1] = 5
ls
```

```python
arrary2[1][2] = 10
arrary2
```

```python
# 브로드 캐스팅
arrary2[0] = 0
arrary2
```

```python
arrary2[0] = [7, 8, 9]
arrary2
```

```python
# 조건으로 선택
arrary2 > 7
```

```python
idx = arrary2 > 7
arrary2[idx]
```

```python
arrary2[idx] = 100
arrary2
```
