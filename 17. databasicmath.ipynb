# 합집합과 교집합

```python
A1 = {1, 2, 3, 4}
A2 = {2, 4, 6}
A3 = {1, 2, 3}
A4 = {2, 3, 4, 5, 6}
```

```python
A1.union(A2)
```

```python
A2 | A1
```

```python
A3.intersection(A4)
```

```python
A3 & A4
```
