# PCA
- principal component analysis는 주성분 분석이라고도하면 고차원 데이터 집항이 주어졌을 때 고차원 데이터와 가장 비슷
  하면서 더 낮은 차원 데이터를 찾아내는 방법 
  차원축소라고도 한다 더 낮은 차원데 ㅇ데이터값 변화가 더 높은 차원에 디이터값 변화를 설명할 수 있다는 것은 얼핏 보기에
  복잡해 보이는 고차원데이터의 변이를 몇가지 원인으로 설명할 수 있다는 뜻이다.

# 사이킷런의 PCA기능
- n_components : 정수
- fit_transfrom() : 특징행렬을 낮은 차원의 근사행렬로 변환
- inverse_transform() : 변환된 근사행렬을 원래의 차원으로 복귀
- mean_ : 평균 벡터
- components_ : 주성분 벡터

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
N = 10 # 잎의 10송이만 선택
X = iris.data[:N, :] # 모두선택
```

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(n_components=1)
```

```python
x_low = pca1.fit_transform(X)
```

```python
x_low
```

```python
X2 = pca1.inverse_transform(x_low)
```

```python
X2
```
