```python
import tensorflow as tf
```

```python
hello = tf.constant("Hello World")
sess = tf.Session()
print(sess.run(hello))
```

```python
# constant 상수
a = tf.constant(1)
b = tf.constant(2)
c = tf.add(a, b)
sess = tf.Session()
print(sess.run(c))
```

```python
# Variable 변수
a = tf.Variable(5)
b = tf.Variable(3)
c = tf.multiply(a, b)
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
sess.run(c)
```
